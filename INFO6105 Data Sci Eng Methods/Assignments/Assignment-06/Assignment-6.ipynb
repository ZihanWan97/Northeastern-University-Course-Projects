# 1. Build a fair die

As we all know, rolling a die is consistent with **discrete uniform distribution**.  

In this case, the probabilities of the 6 outcomes are equal.(This is a perfectly random event.)



```python
face = [1,2,3,4,5,6]
counts, bins = np.histogram(face, bins= 100)
bins = bins[:-1] + (bins[1] - bins[0])/2
probs = counts/float(counts.sum())
plt.bar(bins, probs)
plt.xlabel('Die Roll Outcome', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.title("Roll a Fair Die with a Uniform Distribution")
plt.show()
```

But, the outcomes of rolling a die in reality are as follows.

At first, we roll 1000 times.

```python
#Generate 1000 random numbers with uniform discrete distribution from 1 to 6
face = randint.rvs(1, 7, size=1000)
plt.hist(face,bins=11)
plt.xlabel('Die Roll Outcome', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title("Roll a Fair Die in Reality-1000 times")
```

Then, we roll 100000 times!  

This looks different from the outcome of rolling 100 times.

```python
#Generate 100000 random samples with uniform discrete distribution from 1 to 6
face = randint.rvs(1, 7, size=100000)
plt.hist(face,bins=11)
plt.xlabel('Die Roll Outcome', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title("Roll a Fair Die in Reality-100000 times")
```

So, in reality, when the number of rolling times is large enough, the probabilities of the six outcomes are nearly equal.

# 2. Build a randomly weighted die 

```python
#Generate a list containing 6 random weights that sum to 1
weight = np.random.dirichlet(np.ones(6),size=1)
weight = list(weight[0])
weight
```

```python
face = [1,2,3,4,5,6]
#Generate 10000 random samples from the list 'face' with weight 'p' 
outcome = np.random.choice(face, 10000, p=weight)
plt.hist(outcome,bins=11)
plt.xlabel('Die Roll Outcome', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title("Roll a Weighted Die")
```

# 3.Build a weighted die that matches a gaussian distribution

```python
face = [1,2,3,4,5,6]
#Generate 100 random samples from the list 'face' with weight 'p' follows Gaussian distribution
outcome = np.random.choice(face, 100, p=[0.023,0.136,0.341,0.341,0.136,0.023]) 
plt.hist(outcome,bins=11)
plt.xlabel('Die Roll Outcome', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title("Roll a Weighted Die with a Gaussian distribution")
```

```python
#Calculate the mean of the actual outcomes
actual_mean = outcome.mean()
theoretical_mean = 3.5
#Calculate the mean error
mean_err = actual_mean - theoretical_mean
mean_err
```

```python
#Calculate the standard deviation of the actual outcomes
actual_std = outcome.std()
theoretical_std = 1
#Calculate the std error
std_err = actual_std - theoretical_std
std_err
```
